### 3. Explore the data to gain insights

Etter at dataen er lastet ned og zip-filene er pakket ut, står vi igjen med 2 mapper, en med testdata og en med treningsdata.


```python
import os
import glob
train_dir = (str(DATA/'train'))
test_dir = (str(DATA/'test'))
```

Sorterer dataen...

```python
train_cats_arr = glob.glob(os.path.join(train_dir, 'cat*'))
train_dogs_arr = glob.glob(os.path.join(train_dir, 'dog*'))
train_data_arr = glob.glob(os.path.join(train_dir, '*'))
test_data_arr = glob.glob(os.path.join(test_dir, '*'))
```

Lager en tabell av navnene(labels) til treningsdataen

Cat = 0, Dog = 1

```python
labels = []

for i in train_data_arr:
    if 'cat' in i:
#         labels.append('Cat')
        labels.append(0)
    else:
        labels.append(1)
#         labels.append('Dog')
```

```python
sns.countplot(labels, palette="Set1")
plt.ylabel("Antall")
plt.title("Cats or Dogs")
plt.show()
```

Ser på noen av bildene

```python
np.random.seed(96)
test_pics = glob.glob(os.path.join(test_dir, '*'))

fig = plt.figure(figsize=(16,16))
for x in range(1, 9):
    img_nr = np.random.randint(12500)
#    img_nr = np.random.randint(25) #Fjerne denne når ferdig
    img = mpimg.imread(test_data_arr[img_nr])
#    img = mpimg.imread(train_cats_arr[img_nr])
    fig.add_subplot(4, 4, x)
    plt.axis('off')
    plt.imshow(img)
    
plt.show()
```

### 4. Prepare the data to better expose the underlying data patterns to Machine Learning algorithms

Vi ser at bildene har forskjellige størrelser. Når vi skal jobbe med de senere og bruke modellen vår er det viktig at alle bildene er samme størrelse.

Endrer størrelsen på bildene.

```python
import cv2
#Legg til progressbar? tqdm
IMAGE_SIZE = 65
train_data = []

for i in train_data_arr:
    img = cv2.imread(i)
    new_img = cv2.resize(img, dsize=(IMAGE_SIZE, IMAGE_SIZE), interpolation=cv2.INTER_CUBIC)
    train_data.append(new_img)
```

```python
b,g,r = cv2.split(train_data[2])
img_show = cv2.merge((r,g,b))
plt.imshow(img_show)
plt.show()
```

Gjør dataen klar for Keras-modellen. 

Keras krever at bildene og labels-ene er Numpy-array. Krever i tilegg også at bildene har "4 dimensjoner" på følgende format: (Antall, ROWS, COLS, CHANNELS), i vårt tilfelle skal CHANNELS=3 siden vi håndterer RGB-farge bilder.

```python
X = []
y = []

for i in range(len(train_data)):
    X.append(train_data[i])
    y.append(labels[i])
   
#X = np.array(X)
X = np.array(X).reshape(len(train_data), IMAGE_SIZE, IMAGE_SIZE, 3)
#X = np.array(X).reshape(-1, IMAGE_SIZE, IMAGE_SIZE, 3)
y = np.array(y)
```

Bruker Scikit-learn sin train_test_split metode for å sette av testdata. 20% av dataen setter vi som testdata

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 96)
```

```python
b,g,r = cv2.split(X_train[200])
img_show = cv2.merge((r,g,b))
plt.imshow(img_show)
plt.show()
```

### 5. Explore many different models and short-list the best ones

Nå skal vi lage modellen vår.

Vi bruker Keras sin Sequential model. 

"The Sequential model is a linear stack of layers." - https://keras.io/getting-started/sequential-model-guide/ 

```python
from keras import layers, models, optimizers
from keras import backend as K
from keras import optimizers
```

```python
optimizer_1 = optimizers.RMSprop(lr=0.001, rho=0.9, epsilon=None, decay=0.0)

model = models.Sequential()

model.add(layers.Conv2D(32, (3, 3), input_shape=(IMAGE_SIZE, IMAGE_SIZE, 3)))
model.add(layers.Activation('sigmoid'))
model.add(layers.MaxPooling2D(pool_size=(2, 2)))

model.add(layers.Conv2D(32, (3, 3)))
model.add(layers.Activation('sigmoid'))
model.add(layers.MaxPooling2D(pool_size=(2, 2)))

model.add(layers.Conv2D(64, (3, 3)))
model.add(layers.Activation('sigmoid'))
model.add(layers.MaxPooling2D(pool_size=(2, 2)))

model.add(layers.Flatten())
model.add(layers.Dense(64))
model.add(layers.Activation('sigmoid'))
model.add(layers.Dropout(0.5))
model.add(layers.Dense(1))
model.add(layers.Activation('sigmoid'))

model.compile(loss='binary_crossentropy', optimizer=optimizer_1, metrics=['accuracy'])
    
model.summary()
```

Trener modellen

```python
history = model.fit(X_train, y_train, batch_size=100, epochs=10, validation_split = 0.25, shuffle = True)
```

Nå har vi en modell som gir oss 75% accuracy, et ok resultat men kunne vært bedre.

```python
# https://keras.io/visualization/

plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()
```

Ser på grafene at modellen blir mer og mer nøyaktig jo lengre vi trener den.

Ser på at et par eksempler hvor modellen er i bruk.

```python
predictions = model.predict(X_test, verbose=0)
```

```python
for i in range(0,10):
    if predictions[i, 0] >= 0.5: 
        print('{:.2%} sikker på at dette er en hund'.format(predictions[i][0]))
    else: 
        print('{:.2%} sikker på at dette er en katt'.format(1-predictions[i][0]))
        
    plt.imshow(X_test[i])
    plt.show()
```

Ser at modellen feilslår på noen av eksemplene.

```python
#Lagrer modellen
model.save('catdog_model')
```

### 6. Fine-tune your models and combine them into a great solution

Vi har allerede finjustert modellen vår ganske mye, de første gangene vi kjørte modellen vår fikk vi veldig dårlige resultat på 50%. Det kan tolkes at modellen bare gjettet. 

Dette løste vi med å blant annet øke antall epochs, opprinnelig hadde vi bare 1 epoch, noe som var altfor lite. Prøvde også forskjellige Batch_size-verdier, og endte opp med en verdi på 100 tilslutt.

Fikk også dårlige resultater når Activation var lik relu - "model.add(layers.Activation('relu'))", endret den til sigmoid og fikk mye bedre resultater.

Det er nok mulig å finjustere modellen og få den enda mer nøyaktig. 

Vi kan prøve å lære oss enda mer om Keras, og teknikker for finjustering og å gjøre modellen enda mer effektiv.

Det går også an å prøve å kombinere flere modeller, evt undersøke andre "bedre" modeller.

Det finnes også en funksjon i Keras som heter EarlyStopping, den kan brukes for å forhindre at modellen vil 'overfit'-e, planla å legge den til, men ble litt glemt...
